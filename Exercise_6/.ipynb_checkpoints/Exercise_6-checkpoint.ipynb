# 1 Solving large linera system of equations

In the data directory you will find 5 different linear system of equations of the form 

$$A\vec{x} = \vec{b}$$

For each of them there is a file named "LSEi_A.dat" with the information of the matrix $A$ and a file "LSEi_b.dat" with the information of the vector $\vec{b}$ (with $i = 1,\cdots,5$)

* Write a script that can read in $A$ and $\vec{b}$ of each LSE and allows you tu inspect them. How biare they ?. Can they be solved? (Note that to answer this question you will need to calculate the determinan of each matrix)


```python
path = "data/"
num_matrix = 5
A = [loadtxt(path+"LSE"+str(i+1)+"_A.dat") for i in range(num_matrix)]
b = [loadtxt(path+"LSE"+str(i+1)+"_b.dat") for i in range(num_matrix)]
det = [slogdet(A[i])[0]*slogdet(A[i])[1] for i in range(num_matrix)]

print("Size of Each Matrix :")
for i in range(num_matrix): print("A"+str(i),"-->", shape(A[i]))

print("Determinant of A1 = ",det[0])
print("Determinant of A2 = ",det[1])
print("Determinant of A3 = ",det[2])
print("Determinant of A4 = ",det[3])
print("Determinant of A5 = ",det[4])
```

* Use Gauss elimination and back-sustition to solve he problems.

You may test the code on a simple problem, say $3\times3$ LSE that we present in class. Then run it on the 5 LSEs in the data directory . Use timing (you may need to search how to do this with python) to keep track of how long it takes.



```python
def Gauss(M,b):
    n = len(M)
    for i in range(n):
        b[i] /= M[i,i];M[i] /= M[i,i] 
        for j in range(i+1,n):
            b[j] -= M[j,i]*b[i]; M[j] -= M[j,i]*M[i]
    
    x = zeros(n)
    x[n-1] = b[n-1]
    
    for k in range(n-2,-1,-1):
        Sum = 0.0 
        for j in range(k+1,n):
            Sum -= M[k,j]*x[j]
            x[k] = b[k] +Sum
    return x
```

We use the example used in class

```python

n = 3
M = zeros((n,n))
B = zeros(n)
M[0,0] = 5; M[0,1] = 3; M[0,2] = 4;B[0] = 3
M[1,0] = 2; M[1,1] = 1; M[1,2] = 5;B[1] = 4
M[2,0] = 5; M[2,1] = 4; M[2,2] = 1;B[2] = 2

    
Gauss(M,B)
```

We measure the execution time for each method to compare 

```python
t = []
T = []

for i in range(5):
    to = time.clock()
    Gauss(A[i],b[i])
    tf = time.clock()
    t.append(tf-to)
    
    to = time.clock()
    solve(A[i],b[i])
    tf = time.clock()
    T.append(tf-to)
```

```python
for i in range(5):
    print("A"+str(i+1)+" own method = ",t[i],"numpy routine ",T[i])
```

```python
xaxis = [len(A[i]) for i in range(5)]
plt.plot(xaxis,t,"bo",label = "own method")
plt.plot(xaxis,T,"ro",label = "numpy")
plt.title("Execution time",fontsize = 15)
plt.xlabel("Matrix size",fontsize = 12)
plt.ylabel("Time [s]",fontsize = 12)
plt.grid(True)
plt.legend()
```
